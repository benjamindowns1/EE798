```python
import numpy as np
import pandas as pd


def evaluate():
    # Input the csv file
    """
    Sample evaluation function
    Don't modify this function
    """
    df = pd.read_csv(r"C:\Users\user\Downloads\1f919cc3e58d50718991255baf55c2cade242853d0ee9f149cb7c17da1c37f4f_sample_test (1)\sample_input.csv")
     
    actual_close = np.loadtxt(r"C:\Users\user\Downloads\1f919cc3e58d50718991255baf55c2cade242853d0ee9f149cb7c17da1c37f4f_sample_test (1)\sample_close.txt")
    
    pred_close = predict_func(df)
    
    # Calculation of squared_error
    actual_close = np.array(actual_close)
    pred_close = np.array(pred_close)
    mean_square_error = np.mean(np.square(actual_close-pred_close))


    pred_prev = [df['Close'].iloc[-1]]
    pred_prev.append(pred_close[0])
    pred_curr = pred_close
    
    actual_prev = [df['Close'].iloc[-1]]
    actual_prev.append(actual_close[0])
    actual_curr = actual_close

    # Calculation of directional_accuracy
    pred_dir = np.array(pred_curr)-np.array(pred_prev)
    actual_dir = np.array(actual_curr)-np.array(actual_prev)
    dir_accuracy = np.mean((pred_dir*actual_dir)>0)*100

    print(f'Mean Square Error: {mean_square_error:.6f}\nDirectional Accuracy: {dir_accuracy:.1f}')
    

def predict_func(data):
    """
    Modify this function to predict closing prices for next 2 samples.
    Take care of null values in the sample_input.csv file which are listed as NAN in the dataframe passed to you 
    Args:
        data (pandas Dataframe): contains the 50 continuous time series values for a stock index

    Returns:
        list (2 values): your prediction for closing price of next 2 samples
    """
    import tensorflow as tf
    from sklearn.preprocessing import MinMaxScaler
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, GRU, Dropout



    # Create a new dataframe with only the 'Close column
    closing_prices = data.Close

    closing_prices=closing_prices.ffill()

    # Convert the dataframe to a numpy array
    dataset = closing_prices.values.reshape(-1, 1)

    # Normalizing Data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset)

#     # Extract 'Close' column
#     close_prices = data.Close
#     # Deal with missing values
#     close_prices = close_prices.ffill()
#     close_prices = close_prices.values.reshape(-1, 1)

#     # Normalize
#     scaler = MinMaxScaler()
#     scaled_data = scaler.fit_transform(close_prices)


# Preparing Dataset
    def prepare_prediction_data(data, stepping_time):
        x_predict = []
        for i in range(len(data) - stepping_time):
            x_predict.append(data[i:(i + stepping_time), 0])
        return np.array(x_predict)


    stepping_time = 15
    x_predict = prepare_prediction_data(scaled_data, stepping_time)
    
    tf.keras.backend.clear_session()
    model = Sequential()
    model.add(GRU(32, return_sequences=True, input_shape=(stepping_time, 1)))
    model.add(GRU(32, return_sequences=False))
    model.add(Dropout(0.10))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    model.load_weights(r'C:\Users\user\EE798\MODEL3.h5')

    # Predict future close prices
    predicted_close_prices = model.predict(x_predict)

    predicted_close_prices = scaler.inverse_transform(predicted_close_prices)
    # print('Predicted Close Price for Next Day:', predicted_close_prices[-2])
    # print('Predicted Close Price for Day After Next:', predicted_close_prices[-1])
    return [predicted_close_prices[-2], predicted_close_prices[-1]]
    

if __name__== "__main__":
    evaluate()
```
